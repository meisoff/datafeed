```python
import json
import codecs

data = json.load(codecs.open('/content/drive/MyDrive/Study/SPB_res.json', 'r', 'utf-8-sig'))
```

```python
list_parse = [] # Создаем список для записи url с именами для парсинга сайтов
list_site = [] # Создаем список url для проверки
```

```python
# Парсим инфу с JSON файла
for item in data:
  el = {
      "name": None,
      "url": None
  }
  el["name"] = item['name_ex']['primary']
  
  for item in item['contact_groups'][0]['contacts']: # Получаем ссылку сайта
    if item["type"] == 'website':
      el["url"] = item['text']
  
  if el["url"]:
    if el["url"] not in list_site:
        list_parse.append(el)
        list_site.append(el["url"])
```
