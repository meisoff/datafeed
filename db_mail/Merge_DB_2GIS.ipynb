```python
import json
import codecs

file_names = [
    '2gis_list_msw.json',
    '2gis_list_ekb.json',
    '2gis_list_kzn.json',
    '2gis_list_spb.json'
]

data = [json.load(codecs.open(f'/content/drive/MyDrive/Study/{i}', 'r', 'utf-8-sig')) for i in file_names]
```

```python
# Проверяем на совпадения
def verify_array(data):
  result = []

  for list_ in data:
    # Добавляем первую информацию в массив
    if len(result) == 0:
      result = list_
      continue

    # Сравниваем на наличие повторений в массиве
    for item_append in list_:
      flag = True
      for item_res in result:
        if item_append['name'] == item_res['name'] or item_append['url'] == item_res['url']:
          flag = False # Если в объекте есть совпадения по имени или URL, то не добавляем итем
      if flag:
        result.append(item_append)
  return result
```

```python
print(len(verify_array(data)))
```

```python
# Сохраняем в файл

with open("base_sellers.json", "w") as f:
  f.write(str(verify_array(data)))
```
